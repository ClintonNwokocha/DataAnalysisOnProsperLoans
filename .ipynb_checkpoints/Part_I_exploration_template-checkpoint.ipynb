```python
# Selecting features that are relevant to the question to be answereed
columns = ['ListingNumber', 'ListingCreationDate', 'Term', 'LoanStatus', 'ClosedDate', 'EmploymentStatus',
          'CreditScoreRangeLower', 'CreditScoreRangeUpper', 'PublicRecordsLast10Years', 'DebtToIncomeRatio', 'IncomeRange', 
           'IncomeVerifiable', 'CurrentDelinquencies', 'AmountDelinquent', 'StatedMonthlyIncome','LoanCurrentDaysDelinquent', 'LoanOriginationDate', 'LoanOriginalAmount', 'MonthlyLoanPayment']
loan_df = loan[columns]
```

```python
# Check information of the sub dataframe
loan_df.info()
```

```python
# Check for duplicated values in the dataset
loan_df.duplicated().sum()
```

### There are duplicated values which needs to be dropped

```python
# Dropping duplicate values and confirm the duplicates have been dropped
loan_df = loan_df.drop_duplicates()
loan_df.duplicated().sum()
```

### Filling in missing values

```python
# get the mean and mode of variables with missing values 
mode_employment_status = loan_df['EmploymentStatus'].mode()[0]
mean_csrLower = round(loan_df['CreditScoreRangeLower'].mean(), 0)
mean_csrUpper = round(loan_df['CreditScoreRangeUpper'].mean(), 0)
mean_public_recs = round(loan_df['PublicRecordsLast10Years'].mean(), 0)
mean_debt_to_income_ratio = round(loan_df['DebtToIncomeRatio'].mean(), 2)
mean_current_deliq = round(loan_df['CurrentDelinquencies'].mean(), 0)
mean_amount_deliq = round(loan_df['AmountDelinquent'].mean(), 0)
```

```python
#  fill in midding vaalues
loan_df['EmploymentStatus'] = loan_df['EmploymentStatus'].fillna(mode_employment_status)
loan_df['CreditScoreRangeLower'] = loan_df['CreditScoreRangeLower'].fillna(mean_csrLower)
loan_df['CreditScoreRangeUpper'] = loan_df['CreditScoreRangeUpper'].fillna(mean_csrUpper)
loan_df['PublicRecordsLast10Years'] = loan_df['PublicRecordsLast10Years'].fillna(mean_public_recs)
loan_df['DebtToIncomeRatio'] = loan_df['DebtToIncomeRatio'].fillna(mean_debt_to_income_ratio)
loan_df['CurrentDelinquencies'] = loan_df['CurrentDelinquencies'].fillna(mean_current_deliq)
loan_df['AmountDelinquent'] = loan_df['AmountDelinquent'].fillna(mean_amount_deliq)
loan_df.info()
```

```python
# create a column that gives an average credit score value for each entry
loan_df['AverageCreditScore'] = (loan_df['CreditScoreRangeLower'] + loan_df['CreditScoreRangeUpper']) // 2
loan_df.info()
```

```python
loan_df['AmountDelinquent'].value_counts()
```
